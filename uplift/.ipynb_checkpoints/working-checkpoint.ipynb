# Features

```python
df = receipts.merge(customers, on='customer_id', how='left')
df = df.groupby('age').purchase_amt.sum().compute()
df.reset_index().plot.scatter(x='age', y='purchase_amt')
```

```python
df = receipts.merge(customers, on='customer_id', how='left')
df = df.groupby('age').purchase_amt.count().compute()
df.reset_index().plot.scatter(x='age', y='purchase_amt')
```

```python
df = receipts.merge(customers, on='customer_id', how='left')
df = df.groupby(['age_group', 'location']).purchase_amt.count().compute()
```

```python
df = df.reset_index()
df['loc_age'] = df.apply(lambda x: x['location'] + '_' + str(x['age_group']), axis=1)

df.plot.scatter(x='loc_age', y='purchase_amt')
```

```python
receipts_control = receipts[~receipts.customer_id.isin(treatment_group)]
```

```python
receipts_control['week'] = receipts_control['date'] % 7
receipts_control['price'] = receipts_control['purchase_sum'] / receipts_control['purchase_amt']
receipts_control = receipts_control.merge(customers.loc[:, ['customer_id', 'location']],
                                         on='customer_id', how='left')
```

```python
df = receipts_control.groupby(['date', 'location']).purchase_amt.sum().compute()
df = df.reset_index() 
```

```python
fig, ax = plt.subplots(len(locations), 1, figsize=(24, 18))

for i, l in enumerate(locations):
    df_selection = df[df.location==l]
    ax[i].plot(df_selection.date, df_selection.purchase_amt)
```

```python
PRICE = 80
COST_PRICE = 52
FIRST_GRAMM_DISCOUNT = 40
COMMUNICATION_COST = 1

data_config = {
    'extract': [
        {
            'name': 'receipts_basic',
            'args': {'delta': 30, 'date_to': 101}
        },
        {
            'name': 'receipts_basic',
            'args': {'delta': 15, 'date_to': 101}
        },
        {
            'name': 'age_location',
            'args': {}
        },
        {
            'name': 'location_sales',
            'args': {
                'delta': 30,
                'date_to': 101
            }
        },
        {
            'name': 'location_sales',
            'args': {
                'delta': 15,
                'date_to': 101
            }
        },
        {
            'name': 'target_from_campaign',
            'args': {}
        }
    ],

    'transform': [
        {
            'name': 'expression',
            'args': {
                'expression': f"{{d}}['purchase_sum_sum'] - {{d}}['purchase_amt_sum'] * {COST_PRICE} - {{d}}['treatment_flg'] * {COMMUNICATION_COST} - {{d}}['purchase_treatment_flg'] * {FIRST_GRAMM_DISCOUNT}",
                'col_result': 'target_profit'
            }
        },
       
        {
            'name': 'one_hot_encode',
            'args': {
                'cols': ['location']
            }
        },
        
        {
            'name': 'location_relative_consumption',
            'args': {'periods': [15, 30]}
        },
        {
            'name': 'drop_columns',
            'args': {'columns_to_drop': ['location', 'purchase_sum_sum', 'purchase_amt_sum',
                                         'purchase_treatment_flg']}
        },
    ]   
      
}
```

```python
df = compute.extract_features(engine=engine, config=data_config['extract'])
```

```python
transform_pipeline = compute.create_pipeline(data_config['transform'])
df = transform_pipeline.fit_transform(df)
```
