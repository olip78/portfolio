```python
import os
import sys
sys.path.append('./uplift-campaign/')

import pandas as pd
import numpy as np
import dask.dataframe as dd

import copy
import json
import pickle

import seaborn as sns
from matplotlib import pyplot as plt
sns.set_theme()
```

```python
from upcampaign.datalib.features import extract, compute
import upcampaign.model_utils as mu
import upcampaign.datalib.connection as connection
```

```python
import lightgbm as lgb
import causalml.metrics as cmetrics
import causalml.inference.meta.xlearner as xlearner
import causalml.inference.meta.rlearner as rlearner

import pylift

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import sklearn.pipeline as skpipe
```

```python
customers = pd.read_csv('./data/customers.csv', index_col=0)
campaigns = pd.read_csv('./data/campaigns.csv', index_col=0)
receipts = dd.read_parquet('./data/receipts.parquet')

engine = connection.Engine(tables={
    'customers': customers,
    'campaigns': campaigns,
    'receipts': receipts
})
```

# Getting training data set

```python
# training config

PRICE = 80
COST_PRICE = 52
FIRST_GRAMM_DISCOUNT = 40
COMMUNICATION_COST = 1

d = 101
periods = [15, 30, 60, 90, 102, 7]

extract = [{'name': 'age_location', 'args': {}},        
           {'name': 'target_from_campaign','args': {}}]

for delta in periods:
    extract.append({'name': 'receipts_basic',
                    'args': {'delta': delta, 'date_to': d}})
    extract.append({'name': 'location_age_sales',
                    'args': {'delta': delta, 'date_to': d}})
_
data_config = {
    'extract': extract,

    'transform': [
        {
            'name': 'expression',
            'args': {
                'expression': f"{{d}}['purchase_sum_sum'] - {{d}}['purchase_amt_sum'] * {COST_PRICE} - {{d}}['treatment_flg'] * {COMMUNICATION_COST} - {{d}}['purchase_treatment_flg']*{{d}}['discount_flag_max']* {FIRST_GRAMM_DISCOUNT}",
                'col_result': 'target_profit'
            }
        },
       
        {
            'name': 'one_hot_encode',
            'args': {
                'cols': ['location']
            }
        },
        
        {
            'name': 'location_relative_consumption',
            'args': {'periods': periods}
        },
        {
            'name': 'drop_columns',
            'args': {'columns_to_drop': ['location', 'purchase_sum_sum', 'purchase_amt_sum',
                                         'purchase_treatment_flg']}
        },
    ]   
      
}
```

```python
# extracting features
data = compute.extract_features(engine=engine, config=data_config['extract'])

# features transformation
transform_pipeline = compute.create_pipeline(data_config['transform'])
data_transformed_row = transform_pipeline.fit_transform(data)
```

```python
# drop nans
tmp = data_transformed_row.isna().sum()
n_hist = tmp[tmp>0].shape[0]

tmp = data_transformed_row.isna().sum(axis=1)

data_transformed = data_transformed_row.loc[tmp[tmp<n_hist].index, :]
data_transformed = data_transformed.fillna(-999)
```

# Modelling

```python
col_target = 'target_profit'
col_treatment = 'treatment_flg'

cols_features = list(data_transformed.columns)
columns_to_drop = [col_target, col_treatment, 'customer_id', 
                   'discount_flag_max'] + ['location_age'] + ['purchase_amt_count__30d',
                                                              'purchase_amt_count__60d',
                                                             'purchase_amt_count__15d',
                                                             'purchase_amt_count__7d',
                                                              'purchase_amt_mean__7d',
                                                              'purchase_amt_count__102d',
                                                             ]

columns_to_drop = set(columns_to_drop)

for c in cols_features:
    if ('*' in c):
        columns_to_drop.add(c)
    if 'ohe' in c:
        columns_to_drop.add(c)    

for c in columns_to_drop:
    cols_features.remove(c)
    
print(f'n of features: {len(cols_features)}')
```

```python
fit_index, val_index = train_test_split(data_transformed.index, test_size=0.25, random_state=165)
```

#### r learner

```python
model_r_n5 = rlearner.BaseRRegressor(
    learner=lgb.LGBMRegressor(),
    control_name=0,
    n_fold=5,   # 5 best
    random_state=165
)
```

```python
_p_score_const = data_transformed.loc[fit_index, col_treatment].value_counts(normalize=True).loc[1]
_p_scores = pd.Series(data=_p_score_const, index=fit_index)
model_r_n5.fit(
    X=data_transformed.loc[fit_index, cols_features].values,
    treatment=data_transformed.loc[fit_index, col_treatment].values,
    y=data_transformed.loc[fit_index, col_target].values,
    p=_p_scores
)
```

```python
_eval_data = data_transformed.loc[val_index, cols_features + [col_treatment, col_target]]
_eval_data['score'] = model_r_n5.predict(_eval_data[cols_features])
_eval_data = _eval_data[['score', col_treatment, col_target]]
```

```python
cols_features = list(data_transformed.columns)
columns_to_drop = [col_target, col_treatment, 'customer_id', 
                   'discount_flag_max'] + ['location_age'] + ['purchase_amt_count__30d',
                                                              'purchase_amt_count__60d',
                                                             'purchase_amt_count__15d',
                                                             'purchase_amt_count__7d',
                                                              'purchase_amt_mean__7d',
                                                              'purchase_amt_count__102d',
                                                             ]

columns_to_drop = set(columns_to_drop)

for c in cols_features:
    if ('ohe' in c):
        columns_to_drop.add(c)
    if '*' in c:
        columns_to_drop.add(c)    

for c in columns_to_drop:
    cols_features.remove(c)
    
print(f'n of features: {len(cols_features)}')
```

```python
model_x = xlearner.BaseXRegressor(
    learner=lgb.LGBMRegressor(),
    control_name=0,
)
```

```python
_p_scores = pd.Series(data=_p_score_const, index=val_index)
_eval_data = data_transformed.loc[val_index, cols_features + [col_treatment, col_target]]

_eval_data['score'] = model_x.predict(_eval_data[cols_features], p=_p_scores)
_eval_data = _eval_data[['score', col_treatment, col_target]]
```

#### trees (rf)

```python
from causalml.inference.tree import UpliftRandomForestClassifier
```

```python
n = 100_000
_sample_treated_size = n
_sample_control_size = n
fit_sample_index = data_transformed.loc[fit_index, [col_treatment]].query(f'{col_treatment} == 1').sample(_sample_treated_size).index
fit_sample_index = fit_sample_index.union(
    data.loc[fit_index, [col_treatment]].query(f'{col_treatment} == 0').sample(_sample_control_size).index
)

_sample_treated_size = n
_sample_control_size = n
val_sample_index = data_transformed.loc[val_index, [col_treatment]].query(f'{col_treatment} == 1').sample(_sample_treated_size).index
val_sample_index = val_sample_index.union(
    data.loc[val_index, [col_treatment]].query(f'{col_treatment} == 0').sample(_sample_control_size).index
)
```

```python
upmodel_rf = UpliftRandomForestClassifier(
    control_name='control',
    evaluationFunction='KL',
    n_estimators=20,
    max_depth=3,
    min_samples_leaf=100
)
```

```python
upmodel_rf.fit(
    X=data_transformed.loc[fit_sample_index, cols_features].values,
    treatment=data_transformed.loc[fit_sample_index, col_treatment].apply(lambda t: 'treatment' if t == 1 else 'control').values,
    y=data_transformed.loc[fit_sample_index, col_target].values
)
```

```python
_eval_data = data_transformed.loc[val_sample_index, cols_features + [col_treatment, col_target]]
_eval_data['score'] = upmodel_rf.predict(_eval_data[cols_features])
_eval_data = _eval_data[['score', col_treatment, col_target]]
```
