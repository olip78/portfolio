```python
strata_age_gender = [
          'gender==1 & ( 25 <= age <= 35) ',
          'gender==0 & ( 25 <= age <= 35)',
          'gender==1 & ((age < 25 )|(age > 35))',
          'gender==0 & ((age < 25 )|(age > 35))',
         ]

strata_age_gender_new = [
          'gender==1 & ( 25 <= age <= 35) & (new==0)',
          'gender==0 & ( 25 <= age <= 35) & (new==0)',
          'gender==1 & ((age < 25 )|(age > 35)) & (new==0)',
          'gender==0 & ((age < 25 )|(age > 35)) & (new==0)',
          'new==1'
         ]
```

```python
def strata_summary(strata, weekly_sales=weekly_sales, target = 'sales'):
    counter = 0
    strata_weights = []
    res = []

    N = weekly_sales.shape[0]
    for query in strata:
        tmp = weekly_sales.query(query)
        n = tmp.shape[0]/N
        res.append([tmp[target].mean(), tmp[target].var(), n])
        strata_weights.append(n)
        counter += tmp.shape[0]

    assert counter == weekly_sales.shape[0]

    df = pd.DataFrame(res, columns=['mean', 'var', 'weight'])
    
    return df, strata_weights
```

```python
def get_strata_mu_std(df, strat_column, target):
    f_std = lambda x: np.sqrt(x.var())
    f_mu = lambda x: x.mean()
    df_by_strata = df.groupby(strat_column, as_index=False).agg(mu=(target, 'mean'),
                                                                std=(target, f_std))
    return df_by_strata
```

###  <a id="4">AAB tests check</a>

In order to verify the validity of the statistical tests as well as the CUPED + and/or post stratification approaches, several aa/ab experiments were carried out on historical data:

    - simple AA t tests on sales / cuped data
    - simple AB t tests on sales / cuped data with additional artificial additional effect
    - confident interval AA tests on sales / cuped post-strst. data
    - confident interval AB tests on sales / cuped post-strst. data with additional artificial additional effect
    
experimental results should be consistent with theoretical expectations as well as with the MDV from the previous subsection

```python
all_users = users.user_id.unique()

def make_test_groups(all_users, n=1500):
    groups = np.random.choice(all_users, (n, 3), False)
    groups = groups.T
    return {'test': {'group_a_one': list(groups[0]), 'group_a_two': list(groups[1]), 'group_b': list(groups[2])}}

def ttest_test(groups, ws, target, alternative):
    _, pvalue = ttest_ind(ws[ws.user_id.isin(groups['pilot'])][target['pilot']],
                          ws[ws.user_id.isin(groups['control'])][target['control']],
                          alternative=alternative
                          )
    return pvalue

def ttest_test_cuped(groups, ws, alternative, effect=1.03):
    
    ws['sales_tmp'] = ws['sales']
    ii = ws[ws.user_id.isin(groups['pilot'])].index
    ws.loc[ii, 'sales_tmp'] = effect*ws.loc[ii, 'sales_tmp'].values
    
    theta = calculate_theta(ws.sales_tmp.values, ws.covariate.values) 
    ws['cuped_tmp'] = ws.sales_tmp.values - theta*ws.covariate.values
    
    _, pvalue = ttest_ind(ws[ws.user_id.isin(groups['pilot'])]['cuped_tmp'],
                          ws[ws.user_id.isin(groups['control'])]['cuped_tmp'],
                          alternative=alternative
                          )
    return pvalue

```

```python
# for historical AA tests
effect = 1.03

ws['sales_effect'] = effect*ws.sales.values
```

#### without post-stratification

```python
def test_simple_ttest(target, 
                      ws=ws, all_users=all_users, n=5000, 
                      type_=['AA', '1st type error'], 
                      alternative='two-sided',
                      effect=1.03
                     ):
    """n ttest AA/AB experiments  
    """
    res = []
    pvalues = []
    for _ in range(n):
        request = make_test_groups(all_users)
        groups = {'control': request['test']['group_a_one'], 'pilot': request['test']['group_a_two']}
        
        pvalue = ttest_test(groups, ws, target, alternative=alternative)

        if type_[0] == 'AA':
            res.append(pvalue < 0.05)
        elif type_[0] == 'AB':
            res.append(pvalue > 0.05)  
        pvalues.append(pvalue)

    print(f'{type_[0]} historical ttest {target}, {type_[1]}: {np.mean(res)}')
    return pvalues

def test_simple_ttest_cuped(
                      ws=ws, all_users=all_users, n=5000, 
                      effect=1.03
                     ):
    res = []
    pvalues = []
    for _ in range(n):
        request = make_test_groups(all_users)
        groups = {'control': request['test']['group_a_one'], 'pilot': request['test']['group_a_two']}
        
        pvalue = ttest_test_cuped(groups, ws, alternative='greater', effect=effect)
        res.append(pvalue > 0.05)  
        pvalues.append(pvalue)

    print(f'AB historical ttest cuped, 2. type error: {np.mean(res)}')
    return pvalues
```

#### with post-stratification

```python
from lib.tests import calc_strat_mean_var, run_strat_ttest_two_sides, run_strat_ttest 
```

```python
def test_strat_ttest(target_name, strat_column, strata_weights, test,
                     ws=ws, all_users=all_users, n=5000, effect=1., type_=['AA', '1st type error']
                    ):

    results = []
    for _ in range(n):
        request = make_test_groups(all_users)
        groups = {'control': request['test']['group_a_one'], 
                  'pilot': request['test']['group_a_two']}
        
        ws['sales_tmp'] = ws['sales']
        ii = ws[ws.user_id.isin(groups['pilot'])].index
        ws.loc[ii, 'sales_tmp'] = effect*ws.loc[ii, 'sales_tmp'].values
        target_name_new = 'sales_tmp'
        
        if target_name == 'cuped':
            theta = calculate_theta(ws.sales_tmp.values, ws.covariate.values) 
            ws['cuped_tmp'] = ws.sales_tmp.values - theta*ws.covariate.values
            target_name_new = 'cuped_tmp'
        
        ws_pilot = ws[ws.user_id.isin(groups['pilot'])]
        ws_control = ws[ws.user_id.isin(groups['control'])]
        
        res = test(ws_pilot,
                   ws_control,
                   strat_column, target_name_new,
                   strata_weights,
                  )
        results.append(res)
    
    if type_[0] == 'AA':
        res = np.mean(results)
    elif type_[0] == 'AB':
        res = 1 - np.mean(results)
        
    print(f'{type_[0]} historical conf. interval test {strat_column} {target_name}, {type_[1]}: {res}')
```

```python
target_name = 'sales'

strat_column = 'strata_age_gender'
strata_weights = strata_weights_age_gender
test_strat_ttest(target_name, strat_column, strata_weights, run_strat_ttest_two_sides)

strat_column = 'strata_age_gender_new'
strata_weights = strata_weights_age_gender_new
test_strat_ttest(target_name, strat_column, strata_weights, run_strat_ttest_two_sides)
```
